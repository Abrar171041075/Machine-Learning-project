# Split the dataste into train and test

```python
from sklearn.model_selection import train_test_split as tts
```

```python
# train 70% ; test = 30%
xtrain, xtest, ytrain, ytest = tts(x,y,test_size=.30)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(xtrain, ytrain) #training
```

```python
m = reg.coef_
m
```

```python
c = reg.intercept_
c 
```

```python
x1 = 71.713675
y = m * x1 + c
```

```python
y
```

```python

reg.predict([[71.713675]])
```

```python
pred = reg.predict(xtest)
```

```python
pred
```

```python
df['Predicted Result'] = reg.predict(x)
```

```python
df.head()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse = mean_squared_error(ytest, pred)
```

```python
mse
```

```python
mae = mean_absolute_error(ytest, pred)
```

```python
mae
```

# Best fit line

```python
plt.figure(figsize=(20,10))
plt.scatter(df.Height , df.Weight, marker='^' , color='#0B0EE6') # https://htmlcolorcodes.com/
plt.xlabel('Height', color='black')
plt.ylabel('Weight')
plt.title('Weight plot', color='red')
plt.plot(xtrain , reg.predict(xtrain), color='red')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.xlabel('area', color='black')
plt.ylabel('Rent')
plt.title('Rent plot', color='red')
plt.plot(xtrain , reg.predict(xtrain), color='red')
#plt.scatter(xtrain, ytrain, color='yellow')
plt.scatter(xtrain, reg.predict(xtrain), color='green')
plt.show()
```
