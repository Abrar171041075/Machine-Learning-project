# Ordinal Encoding

```python
df8
```

```python
region = df8.region.unique()
region
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordinal = OrdinalEncoder(categories =[region])
```

```python
encoded_value = ordinal.fit_transform(df8[['region']])
```

```python
encoded_value=pd.DataFrame(encoded_value,columns=['region'])
encoded_value.head()
```

```python
df8= pd.concat([df8.drop('region', axis=1),encoded_value],axis=1)
df8.head()
```
